```python
def GraficoInteractivo(v0, theta, phi, vw_x, vw_y, vw_z, omegax, omegay, omegaz):
    
    def SolucionaSistemaEcuaciones(t, M):
        vx = M[0]
        x = M[1]
        vy = M[2]
        y = M[3]
        vz = M[4]
        z = M[5]

        # Ecuaciones de movimiento
        v = np.sqrt((vx - vw_x)**2 + (vy - vw_y)**2 + (vz - vw_z)**2)
        omega = np.sqrt((omegax)**2 + (omegay)**2 + (omegaz)**2)

        if omega == 0:
            dvx_dt = -k * v * (Cd * (vx - vw_x))
            dx_dt = vx

            dvy_dt = -k * v * (Cd * (vy - vw_y))
            dy_dt = vy

            dvz_dt = -g - k * v * (Cd * (vz - vw_z)) + densidadAire * volumenBalon * g / masa
            dz_dt = vz
        else:
            dvx_dt = -k * v * ( Cd *(vx - vw_x) - CL*((omegay*(vz-vw_z) - omegaz*(vy - vw_y))/omega))
            dx_dt = vx

            dvy_dt = -k * v * (Cd * (vy - vw_y) - CL * ((omegaz*(vx - vw_x) - omegax*(vz-vw_z)) / omega))
            dy_dt = vy

            dvz_dt = -g - k * v * ( Cd *(vz - vw_z) - CL*((omegax*(vy - vw_y) - omegay*(vx - vw_x))/omega)) + densidadAire*volumenBalon*g/masa
            dz_dt = vz

        sistemaEcuaciones = np.array([dvx_dt, dx_dt, dvy_dt, dy_dt, dvz_dt, dz_dt])

        return sistemaEcuaciones
    
    # Parámetros iniciales
    # Condiones geométricas y físicas
    g = 9.81 # m/s^2
    # v0 = 30.0 # m/s
    # theta = 45 # grados
    # phi = 45 # grados
    conversion = np.pi/180

    # Condiciones del balón
    masa = 0.450 # kg
    radio = 0.1114 # m
    volumenBalon = (4*np.pi/3)*radio**3 # m^3
    areaTransversal = np.pi*(radio)**2
    densidadAire = 1.225 # kg/m^3
    k = 0.5*densidadAire*areaTransversal/masa
    Cd = 0.13075 # sin unidades
    CL = 0.25 # sin unidades

    # Condiciones Iniciales
    # Movimiento x (Condiciones iniciales)
    vx0 = v0*np.cos(theta*conversion)*np.cos(phi*conversion)
    x0 = 0

    # Movimiento y (Condiciones iniciales)
    vy0 = v0*np.cos(theta*conversion)*np.sin(phi*conversion)
    y0 = 0

    # Movimiento z (Condiciones iniciales)
    vz0 = v0*np.sin(theta*conversion)
    z0 = 0


    # Condiciones del viento
    # Tomando en cuenta el viento
#     vw_x = 0 # m/s
#     vw_y = 0 # m/s
#     vw_z = 0 # m/s

    # Velocidades iniciales angulares
#     omegax = 0 # rad/s
#     omegay = 0 # rad/s
#     omegaz = 0 # rad/s

    # Inicializando el arreglo que contendrá los tiempos
    t_span = np.array([0, 10])
    tiempos = np.linspace(t_span[0], t_span[1], 100)

    # Creando el arreglo de condiciones iniciales
    # Condiciones Iniciales
    condicionesIniciales = np.array([vx0, x0, vy0, y0, vz0, z0])

    # Ejecusión de la función solve_ivp de la biblioteca scipy.integrate
    solucion = solve_ivp(SolucionaSistemaEcuaciones , t_span, condicionesIniciales, t_eval=tiempos)

    # Extracción de los datos
    # Se extraen los datos, se usan solo los valores positivos
    t = solucion.t; t_nuevo=[]
    vx = solucion.y[0]; vx_nuevo=[]
    x = solucion.y[1]; x_nuevo=[]
    vy = solucion.y[2]; vy_nuevo=[]
    y = solucion.y[3]; y_nuevo=[]
    vz = solucion.y[4]; vz_nuevo=[]
    z = solucion.y[5]; z_nuevo=[]

    for i in range(len(z)):
        if z[i] >= 0:
            y_nuevo.append(y[i])
            t_nuevo.append(t[i])
            vx_nuevo.append(vx[i])
            vy_nuevo.append(vy[i])
            x_nuevo.append(x[i])
            vz_nuevo.append(vz[i])
            z_nuevo.append(z[i])
        else:
            break

    t = t_nuevo
    vx = vx_nuevo
    x = x_nuevo
    vy = vy_nuevo
    y = y_nuevo
    vz = vz_nuevo
    z = z_nuevo

    # Sección de graficación en 3D
    fig = plt.figure(figsize=(8,8))
    ax = fig.gca(projection='3d')

    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    ax.set_zlabel('z (m)')
    ax.set_xlim3d(0,100)
    ax.set_ylim3d(0,100)
    ax.set_zlim3d(0,100)
    ax.invert_xaxis()
    ax.plot(x, y, z, color = 'g', label = 'Con fricción')
    X = np.random.randint(0, 150, (5,5))
    Y = np.random.randint(0, 150, (5,5))
    Z = np.random.randint(0, 150, (5,5))
    ax.quiver(X, Y, Z, vw_x, vw_y, vw_z, length = 1, arrow_length_ratio = 0.2)
    ax.legend(loc='upper left')
    ax.invert_xaxis()
    plt.show()
    
    return
```

```python
y=interactive(GraficoInteractivo, v0=widgets.IntSlider(min=5.0, max=60.0, step=5.0, value=5, description=r'$V_{0} (m/s):$'),
              theta=widgets.IntSlider(min=0.0, max=90.0, step=5.0, value=45.0, description=r'$ \theta (grados):$'),
              phi=widgets.IntSlider(min=0.0, max=90.0, step=5.0, value=45.0, description=r'$ \phi (grados):$'), 
              vw_x=widgets.IntSlider(min=-10.0, max=10.0, step=2.0, value=0.0, description=r'$V_{viento_{x}} (m/s):$'),
              vw_y=widgets.IntSlider(min=-10.0, max=10.0, step=2.0, value=0.0, description=r'$V_{viento_{y}} (m/s):$'),
              vw_z=widgets.IntSlider(min=-10.0, max=10.0, step=2.0, value=0.0, description=r'$V_{viento_{z}} (m/s):$'),
              omegax=widgets.IntSlider(min=-5.0, max=5.0, step=1.0, value=0.0, description=r'$\omega_{x} (rad/s):$'),
              omegay=widgets.IntSlider(min=-5.0, max=5.0, step=1.0, value=0.0, description=r'$\omega_{y} (rad/s):$'),
              omegaz=widgets.IntSlider(min=-5.0, max=5.0, step=1.0, value=0.0, description=r'$\omega_{z} (rad/s):$'),)
display(y)
```
